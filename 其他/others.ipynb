# 整数反转，题号:7

给你一个 32 位的有符号整数 x ，返回将 x 中的数字部分反转后的结果。

如果反转后整数超过 32 位的有符号整数的范围 [−2^31,  2^31 − 1] ，就返回 0。

**假设环境不允许存储 64 位整数（有符号或无符号）。**



```python
# 初解

class Solution:
    def reverse(self, x: int) -> int:
        sign = 1
        if x < 0:
            sign = -1
        
        x_str = str( abs(x) )
        x_inverse_str = x_str[::-1]
        x_inverse = int(x_inverse_str) * sign


        if -2**31 <= x_inverse <= 2**31 -1:
            return x_inverse
        else:
            return 0

'''
问题:
1. 没有注意到位数问题：环境不允许存储64位整数。
2. 应使用算法思维去解决：
    按位来反转，利用地板除来得到输入的末尾位数字并反转
    一边读取一边判断是否超出范围
    注意Python地板除的正负问题
'''
class Solution:
    def reverse(self, x: int) -> int:
        MAX_INT = 2**31 -1
        MIN_INT = -2**31
        # 为了提前判断，即应该在最后一位计算溢出之前做出判断
        MAX_INT10 = MAX_INT // 10
        MIN_INT10 = MIN_INT // 10 + 1
        
        # 两个边界值最后一位的数字
        MAX_POP = MAX_INT % 10
        MIN_POP = -(MIN_INT % 10)
        
        res = 0
        sign = 1
        if x < 0:
            sign = -1

        x = abs(x)
        while x != 0:
            pop = x % 10 # 取最后一位数字
            x = x//10 # 把x挪一位

            '''
            判断数字是否超过边界值:
            1. 当数字还没调转结束，但除了最后一个数字，已经大于了 MAX_INT10, 则不管最后一个数字是什么，都会超出范围
            2. 当除了最后一个数字，前面的数字已经等于了 MAX_INT10，则要比较最后一个POP
            '''
            if res > MAX_INT10


            res = res * 10 + pop # 将末尾的数字乘10，即变成了首部的数字



    return res
```
